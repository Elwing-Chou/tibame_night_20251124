````python
# 五子棋
# 1. 先把棋盤畫出來
# 2. 落子(落到最相鄰的位置)
# 3. 判斷落子是否五子
import pygame as pg

# no one win: False
def check_win(board, i, j):
    check = board[i][j]
    # row
    # count = 0
    # for j_offset in range(min(len(board)-j-1, 4)):
    #     if board[i][j+j_offset+1] == check:
    #         count = count + 1
    # if count == 4:
    #     return True
    # 橫的確認
    count = 0
    offset = 1
    while True:
        check_i, check_j = i, j + offset
        if 0 <= check_i < len(board) and 0 <= check_j < len(board[0]):
            if board[check_i][check_j] == check:
                count = count + 1
            else:
                # 遇到不一樣顏色
                break
        else:
            # 撞到牆了
            break
        offset = offset + 1
    offset = 1
    while True:
        check_i, check_j = i, j - offset
        if 0 <= check_i < len(board) and 0 <= check_j < len(board[0]):
            if board[check_i][check_j] == check:
                count = count + 1
            else:
                break
        else:
            break
        offset = offset + 1
    if count >= 4:
        return True
    # col
    count = 0
    offset = 1
    while True:
        check_i, check_j = i + offset, j
        if 0 <= check_i < len(board) and 0 <= check_j < len(board[0]):
            if board[check_i][check_j] == check:
                count = count + 1
            else:
                # 遇到不一樣顏色
                break
        else:
            # 撞到牆了
            break
        offset = offset + 1
    offset = 1
    while True:
        check_i, check_j = i - offset, j
        if 0 <= check_i < len(board) and 0 <= check_j < len(board[0]):
            if board[check_i][check_j] == check:
                count = count + 1
            else:
                break
        else:
            break
        offset = offset + 1
    if count >= 4:
        return True
    # 又斜
    count = 0
    offset = 1
    while True:
        check_i, check_j = i - offset, j + offset
        if 0 <= check_i < len(board) and 0 <= check_j < len(board[0]):
            if board[check_i][check_j] == check:
                count = count + 1
            else:
                # 遇到不一樣顏色
                break
        else:
            # 撞到牆了
            break
        offset = offset + 1
    offset = 1
    while True:
        check_i, check_j = i + offset, j - offset
        if 0 <= check_i < len(board) and 0 <= check_j < len(board[0]):
            if board[check_i][check_j] == check:
                count = count + 1
            else:
                break
        else:
            break
        offset = offset + 1
    if count >= 4:
        return True
    # 左上又下
    count = 0
    offset = 1
    while True:
        check_i, check_j = i + offset, j + offset
        if 0 <= check_i < len(board) and 0 <= check_j < len(board[0]):
            if board[check_i][check_j] == check:
                count = count + 1
            else:
                # 遇到不一樣顏色
                break
        else:
            # 撞到牆了
            break
        offset = offset + 1
    offset = 1
    while True:
        check_i, check_j = i - offset, j - offset
        if 0 <= check_i < len(board) and 0 <= check_j < len(board[0]):
            if board[check_i][check_j] == check:
                count = count + 1
            else:
                break
        else:
            break
        offset = offset + 1
    if count >= 4:
        return True
    return False

# pygame初始化
pg.init()
# 設定視窗
board_line = 15
game_board = [[0] * board_line for i in range(board_line)]
player1_symbol = 1
player2_symbol = 2

width, height = 640, 640
x_inter, y_inter = width / (board_line+1), height / (board_line+1)
# 產生視窗
screen = pg.display.set_mode([width, height])
# 設定遊戲標題
pg.display.set_caption("五子棋")
# 建立畫布bg
bg = pg.Surface(screen.get_size())
# 把畫布填滿某個顏色
bg.fill([199, 167, 82])
# 開始畫棋盤
board_line_color = [50, 50, 50]
board_line_width = 2
# 橫的
i = 0
while i < board_line:
    pg.draw.line(bg,
                 board_line_color,
                 [x_inter, y_inter*i+y_inter],
                 [width-x_inter, y_inter*i+y_inter],
                 board_line_width)
    i = i + 1
# 直的
i = 0
while i < board_line:
    pg.draw.line(bg,
                 board_line_color,
                 [x_inter*i+x_inter, y_inter],
                 [x_inter*i+x_inter, height-y_inter],
                 board_line_width)
    i = i + 1

# 把某個圖層貼到上一層
screen.blit(bg, [0, 0])
# 對畫面進行更新(才會真的秀出來)
pg.display.update()

game_round = 0
running = True
while running:
    for event in pg.event.get():
        if event.type == pg.MOUSEBUTTONUP:
            mouse_x, mouse_y = pg.mouse.get_pos()
            mouse_x_line, mouse_y_line = round(mouse_x / x_inter), round(mouse_y / y_inter)

            if (1 <= mouse_x_line <= board_line) and (1 <= mouse_y_line <= board_line):
                # 把座標換成list的位置
                pos_i, pos_j = mouse_y_line-1, mouse_x_line-1
                # 還沒被下過
                if game_board[pos_i][pos_j] == 0:
                    real_mouse_x, real_mouse_y = mouse_x_line * x_inter, mouse_y_line * y_inter
                    circle_radius = min(x_inter, y_inter) / 2
                    if game_round % 2 == 0:
                        player_color = [0, 0, 0]
                        player_symbol = player1_symbol
                    else:
                        player_color = [255, 255, 255]
                        player_symbol = player2_symbol
                    game_board[pos_i][pos_j] = player_symbol
                    pg.draw.circle(bg,
                                   player_color,
                                   [real_mouse_x, real_mouse_y],
                                   circle_radius)
                    # 把某個圖層貼到上一層
                    screen.blit(bg, [0, 0])
                    # 對畫面進行更新(才會真的秀出來)
                    pg.display.update()
                    game_round = game_round + 1
                    # check win
                    if check_win(game_board, pos_i, pos_j):
                        running = False

        # 如果收到的事件是按x
        if event.type == pg.QUIT:
            # 迴圈就會變成while False
            running = False
pg.quit()
````

````python
# 我最終版本
# 五子棋
# 1. 先把棋盤畫出來
# 2. 落子(落到最相鄰的位置)
# 3. 判斷落子是否五子
import pygame as pg
# 往一邊數可以屬到多少個依樣的
def count_same(board, i, j, i_offset, j_offset):
    check = board[i][j]
    count = 0
    step = 1
    while True:
        check_i, check_j = i + step * i_offset, j + step * j_offset
        if 0 <= check_i < len(board) and 0 <= check_j < len(board[0]):
            if board[check_i][check_j] == check:
                count = count + 1
            else:
                break
        else:
            break
        step = step + 1
    return count

# no one win: False
def check_win(board, i, j):
    count1 = count_same(board, i, j, 0, 1)
    count2 = count_same(board, i, j, 0, -1)
    if count1 + count2 >= 4:
        return True
    count1 = count_same(board, i, j, -1, 0)
    count2 = count_same(board, i, j, 1, 0)
    if count1 + count2 >= 4:
        return True
    count1 = count_same(board, i, j, -1, 1)
    count2 = count_same(board, i, j, 1, -1)
    if count1 + count2 >= 4:
        return True
    count1 = count_same(board, i, j, 1, 1)
    count2 = count_same(board, i, j, -1, -1)
    if count1 + count2 >= 4:
        return True
    return False

# pygame初始化
pg.init()
# 設定視窗
board_line = 15
game_board = [[0] * board_line for i in range(board_line)]
player1_symbol = 1
player2_symbol = 2

width, height = 640, 640
x_inter, y_inter = width / (board_line+1), height / (board_line+1)
# 產生視窗
screen = pg.display.set_mode([width, height])
# 設定遊戲標題
pg.display.set_caption("五子棋")
# 建立畫布bg
bg = pg.Surface(screen.get_size())
# 把畫布填滿某個顏色
bg.fill([199, 167, 82])
# 開始畫棋盤
board_line_color = [50, 50, 50]
board_line_width = 2
# 橫的
i = 0
while i < board_line:
    pg.draw.line(bg,
                 board_line_color,
                 [x_inter, y_inter*i+y_inter],
                 [width-x_inter, y_inter*i+y_inter],
                 board_line_width)
    i = i + 1
# 直的
i = 0
while i < board_line:
    pg.draw.line(bg,
                 board_line_color,
                 [x_inter*i+x_inter, y_inter],
                 [x_inter*i+x_inter, height-y_inter],
                 board_line_width)
    i = i + 1

# 把某個圖層貼到上一層
screen.blit(bg, [0, 0])
# 對畫面進行更新(才會真的秀出來)
pg.display.update()

game_round = 0
running = True
while running:
    for event in pg.event.get():
        if event.type == pg.MOUSEBUTTONUP:
            mouse_x, mouse_y = pg.mouse.get_pos()
            mouse_x_line, mouse_y_line = round(mouse_x / x_inter), round(mouse_y / y_inter)

            if (1 <= mouse_x_line <= board_line) and (1 <= mouse_y_line <= board_line):
                # 把座標換成list的位置
                pos_i, pos_j = mouse_y_line-1, mouse_x_line-1
                # 還沒被下過
                if game_board[pos_i][pos_j] == 0:
                    real_mouse_x, real_mouse_y = mouse_x_line * x_inter, mouse_y_line * y_inter
                    circle_radius = min(x_inter, y_inter) / 2
                    if game_round % 2 == 0:
                        player_color = [0, 0, 0]
                        player_symbol = player1_symbol
                    else:
                        player_color = [255, 255, 255]
                        player_symbol = player2_symbol
                    game_board[pos_i][pos_j] = player_symbol
                    pg.draw.circle(bg,
                                   player_color,
                                   [real_mouse_x, real_mouse_y],
                                   circle_radius)
                    # 把某個圖層貼到上一層
                    screen.blit(bg, [0, 0])
                    # 對畫面進行更新(才會真的秀出來)
                    pg.display.update()
                    game_round = game_round + 1
                    # check win
                    if check_win(game_board, pos_i, pos_j):
                        running = False

        # 如果收到的事件是按x
        if event.type == pg.QUIT:
            # 迴圈就會變成while False
            running = False
pg.quit()
````

````python
# 10 * 8 的長方形
# oxoxoxox
# xoxoxoxo
# oxoxoxox
# 訣竅: 從裡面開始, 從簡單的解決
# 1. 先能夠印出一行 oooooooo->oxoxoxox
# 2. 擴充成為十航 oxoxoxox
# 3. 航也要交替
symbols = ["o", "x"]

for i in range(10):
    symbols_start = i % len(symbols)
    line = ""
    for j in range(8):
        symbol_idx = (symbols_start + j) % len(symbols)
        line = line + symbols[symbol_idx]
    print(line)
````
