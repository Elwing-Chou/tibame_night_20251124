```python
# 五子棋
# 1. 先把棋盤畫出來
# 2. 落子(落到最相鄰的位置)
# 3. 判斷落子是否五子
import pygame as pg

#pygame初始化
pg.init()
#設定視窗
width, height = 640, 640
x_inter, y_inter = width / 16, height / 16
# 產生視窗
screen = pg.display.set_mode([width, height])
# 設定遊戲標題
pg.display.set_caption("五子棋")
# 建立畫布bg
bg = pg.Surface(screen.get_size())
# 把畫布填滿某個顏色
bg.fill([199, 167, 82])
# 開始畫棋盤
board_line_color = [50, 50, 50]
board_line_width = 2
# 橫的
i = 0
while i < 15:
    pg.draw.line(bg,
                 board_line_color,
                 [x_inter, y_inter*i+y_inter],
                 [width-x_inter, y_inter*i+y_inter],
                 board_line_width)
    i = i + 1
# 直的
i = 0
while i < 15:
    pg.draw.line(bg,
                 board_line_color,
                 [x_inter*i+x_inter, y_inter],
                 [x_inter*i+x_inter, height-y_inter],
                 board_line_width)
    i = i + 1

# 把某個圖層貼到上一層
screen.blit(bg, [0, 0])
# 對畫面進行更新(才會真的秀出來)
pg.display.update()

running = True
while running:
    for event in pg.event.get():
        # 如果收到的事件是按x
        if event.type == pg.QUIT:
            # 迴圈就會變成while False
            running = False
pg.quit()
```
