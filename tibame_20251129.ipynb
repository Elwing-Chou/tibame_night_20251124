```python
# 五子棋
# 1. 先把棋盤畫出來
# 2. 落子(落到最相鄰的位置)
# 3. 判斷落子是否五子
import pygame as pg

# pygame初始化
pg.init()
# 設定視窗
board_line = 15
width, height = 640, 640
x_inter, y_inter = width / (board_line+1), height / (board_line+1)
# 產生視窗
screen = pg.display.set_mode([width, height])
# 設定遊戲標題
pg.display.set_caption("五子棋")
# 建立畫布bg
bg = pg.Surface(screen.get_size())
# 把畫布填滿某個顏色
bg.fill([199, 167, 82])
# 開始畫棋盤
board_line_color = [50, 50, 50]
board_line_width = 2
# 橫的
i = 0
while i < board_line:
    pg.draw.line(bg,
                 board_line_color,
                 [x_inter, y_inter*i+y_inter],
                 [width-x_inter, y_inter*i+y_inter],
                 board_line_width)
    i = i + 1
# 直的
i = 0
while i < board_line:
    pg.draw.line(bg,
                 board_line_color,
                 [x_inter*i+x_inter, y_inter],
                 [x_inter*i+x_inter, height-y_inter],
                 board_line_width)
    i = i + 1

# 把某個圖層貼到上一層
screen.blit(bg, [0, 0])
# 對畫面進行更新(才會真的秀出來)
pg.display.update()

game_round = 0
running = True
while running:
    for event in pg.event.get():
        if event.type == pg.MOUSEBUTTONUP:
            mouse_x, mouse_y = pg.mouse.get_pos()
            mouse_x_line, mouse_y_line = round(mouse_x / x_inter), round(mouse_y / y_inter)

            if (1 <= mouse_x_line <= board_line) and (1 <= mouse_y_line <= board_line):
                real_mouse_x, real_mouse_y = mouse_x_line * x_inter, mouse_y_line * y_inter
                circle_radius = min(x_inter, y_inter) / 2
                if game_round % 2 == 0:
                    player_color = [0, 0, 0]
                else:
                    player_color = [255, 255, 255]
                pg.draw.circle(bg,
                               player_color,
                               [real_mouse_x, real_mouse_y],
                               circle_radius)
                # 把某個圖層貼到上一層
                screen.blit(bg, [0, 0])
                # 對畫面進行更新(才會真的秀出來)
                pg.display.update()
                game_round = game_round + 1

        # 如果收到的事件是按x
        if event.type == pg.QUIT:
            # 迴圈就會變成while False
            running = False
pg.quit()
```

```python
# 10 * 8 的長方形
# oxoxoxox
# xoxoxoxo
# oxoxoxox
# 訣竅: 從裡面開始, 從簡單的解決
# 1. 先能夠印出一行 oooooooo->oxoxoxox
# 2. 擴充成為十航 oxoxoxox
# 3. 航也要交替
symbols = ["o", "x"]

for i in range(10):
    symbols_start = i % len(symbols)
    line = ""
    for j in range(8):
        symbol_idx = (symbols_start + j) % len(symbols)
        line = line + symbols[symbol_idx]
    print(line)
```

```python
# [
#   [1, 2, 3, 4, 5],
#   [10, 9, 8, 7, 6],
#   [11, 12, 13, 14, 15],
#   [20, 19, 18, 17, 16]
# ]
# 簡化
width, height = 5, 10
a = [[0] * width for i in range(height)]

for i in range(height):
    for j in range(width):
        if i % 2 == 0:
            base = 5 * i + 1
            a[i][j] = base + j
        else:
            base = 5 * i + 5
            a[i][j] = base - j
a
```
