````python
# BeautifulSoup4
import bs4 as bs
import urllib.request as req

url = "https://tabelog.com/tw/tokyo/rstLst/cake/1/?SrtT=rt"
resp = req.urlopen(url)
content = resp.read()
html = bs.BeautifulSoup(content)

# find find_all (區塊html名稱, 篩選條件)
target = html.find("a", {"class":"list-rst__rst-name-target"})
# 萃取東西
# 1. 特殊屬性(字典): 區塊[屬姓名]
# 型態可以擁有兩種東西 1. 專屬功能(f.read()) 2. 專屬值(人.身高)
# 2. 顯示內容: 區塊.text   區塊.get_text()
print(target["href"])
print(target.text)
print(target.get_text())
````
