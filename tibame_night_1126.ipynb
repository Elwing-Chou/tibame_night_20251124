```python
def is_leap_year(year):
    ans = False
    if year % 400 == 0:
        ans = True
    elif year % 100 == 0:
        ans = False
    elif year % 4 == 0:
        ans = True
    else:
        ans = False
    return ans
```

```python
def is_date_valid(date_checked):
  day = date_checked % 100
  remain = date_checked // 100
  month = remain % 100
  year = remain // 100
  # print(year, month, day)

  if year > 0:
    if 1<= month <= 12:
      day_per_month = [31,28,31,30,31,30,31,31,30,31,30,31]
      day_allowed = day_per_month[month - 1]
      if month == 2 and is_leap_year(year) == True:
        day_allowed = day_allowed + 1
      if 1 <= day <= day_allowed:
        return True
      else:
        return False
    else:
      return False
  else:
    return False
```

```python
def thought_2(start=4499, max_count=10):
  ans = []
  count = 0
  while True:
    d = start * start
    if is_date_valid(d):
      ans.append(d)
      count = count + 1
    if count == max_count:
      break
    start = start + 1
  return ans
```

```python
def thought_1(y=2024, m=10, d=1, max_count=10):
  ans = []
  start = dt.datetime(y, m, d)
  dif = dt.timedelta(days=1)
  count = 0
  while True:
    start_str = start.strftime("%Y%m%d")
    start_int = int(start_str)
    start_int_root = start_int ** 0.5
    if abs(round(start_int_root) - start_int_root) < 0.000001:
      # print(start_int)
      ans.append(start_int)
      count = count + 1
    if count == max_count:
      break
    start = start + dif
  return ans
```

```python
import time

count = 10

start = time.time()
thought_1(max_count=count)
end = time.time()
dif1 = end - start
print("想法1:", dif1)

start = time.time()
thought_2(max_count=count)
end = time.time()
dif2 = end - start
print("想法2:", dif2)

ratio = dif1 / dif2
print("差了多少倍:", ratio)
```

```python

import pygame as pg

#pygame初始化
pg.init()
#設定視窗
width, height = 640, 640
# 產生視窗
screen = pg.display.set_mode([width, height])
# 設定遊戲標題
pg.display.set_caption("五子棋")
# 建立畫布bg
bg = pg.Surface(screen.get_size())
# 把畫布填滿某個顏色
bg.fill([199, 167, 82])

# 把某個圖層貼到上一層
screen.blit(bg, [0, 0])
# 對畫面進行更新(才會真的秀出來)
pg.display.update()

running = True
while running:
    for event in pg.event.get():
        # 如果收到的事件是按x
        if event.type == pg.QUIT:
            # 迴圈就會變成while False
            running = False
pg.quit()
```
