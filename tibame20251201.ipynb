````python
import json
import urllib.request as req

url = "https://api.gamer.com.tw/anime/v1/danmu.php?videoSn=36632&geo=TW%2CHK"
f = req.urlopen(url)
content = f.read()
# print(type(content))
content_converted = json.loads(content)
# print(type(content_converted))
danmus = content_converted["data"]["danmu"]

f = open("danmu.csv", "w", encoding="utf-8")
f.write("id,text\n")
for d in danmus:
    d_uid = d["userid"]
    d_text = d["text"]
    print(d_uid, d_text)
    line_pat = '"{}","{}"\n'
    line = line_pat.format(d_uid, d_text)
    f.write(line)
f.close()
````
