dict: 一個複雜單位

list+dict: JSON格式
展覽清單 = [
    {
        "主辦人":[
            {人}, {人}, {人}
        ]
    }
]

````python
import json
import urllib.request as req

url = "https://api.gamer.com.tw/anime/v1/danmu.php?videoSn=36632&geo=TW%2CHK"
f = req.urlopen(url)
content = f.read()
# print(type(content))
content_converted = json.loads(content)
# print(type(content_converted))
danmus = content_converted["data"]["danmu"]
for d in danmus:
    d_uid = d["userid"]
    d_text = d["text"]
    print(d_uid, d_text)
````
