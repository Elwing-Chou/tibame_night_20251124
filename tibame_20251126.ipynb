```python
def is_leap_year(year):
    ans = False
    if year % 400 == 0:
        ans = True
    elif year % 100 == 0:
        ans = False
    elif year % 4 == 0:
        ans = True
    else:
        ans = False
    return ans
```

```python
def is_date_valid(date_checked):
    day = date_checked % 100
    remain = date_checked // 100
    month = remain % 100
    year = remain // 100

    if year > 0:
        if 1 <= month <= 12:
            day_per_month = [31, 28, 31, 30, 31, 30, 31, 31, 30, 31, 30, 31]
            day_allowed = day_per_month[month-1]
            if month == 2 and is_leap_year(year) == True:
                day_allowed = day_allowed + 1
            if 1 <= day <= day_allowed:
                return True
            else:
                return False
        else:
            return False
    else:
        return False
```
